# Regression

Regression is the task of training a machine learning model to estimate a continuous variable's value. In other words, we are trying to find a function that will act similar to a series of points, and could be described as an approximation of the function of those points.

Some common use-cases of Regression:
* Stock price forecasting
* Housing prices
* Credit Scores


Tonight, we will be covering the most basic form of regression, Linear Regression. Linear Regression is the most basic form of Regression and involves finding a linear relationship (y = mx + b) between the independent variables (the data we are inputting) and the dependent variable (the data we want out of the model).

But how can we find a line that best describes this curve? For linear regression, we can use the Least Squares Estimate. The equation to calculate the Least Squares Estimate is simple.

For $y = mx + b$:

$m = \frac{\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})}{\sum_{i=1}^{n}(X_i - \bar{X})^2}$

$b = \bar{Y} - m\bar{X}$

Where $\bar{X}$ and $\bar{Y}$ are the average values of the $X$ and $Y$ variables.

How good is this though? We need some way to evaluate this! For linear regression, two good metrics are **Mean Square Error** and **$R^2$**, also known as **Goodness of Fit**.

### Mean Square Error

Before we can talk about mean square error, we must talk about mean absolute error. Mean absolute error is simply the average difference between the true value of an output versus the predicted value:

$Error = \frac{1}{N}\sum|y-\hat{y}|$

The issue with linear error is that it does not greatly punish large differences. While large differences are noticed, a difference of 4 and 5 is quite similar, when we may not want it to be. Therefore, we *square* the values.

$Error = \frac{1}{N}\sum(y-\hat{y})^2$

This increases the error for any large differences even further, and pushes models to minimize the error as much as possible.

### $R^2$

$R^2$, also known as goodness of fit, is a statistic measure that determines the proportion of variance in the dependent variable that can be explained by the independent variable. In other words, how much can we explain the variation in our output based on our model and it's input features. The higher the $R^2$, the better our model describes the data. $R^2$ is calculated by:

$R^2 = 1 - \frac{\sum(y_i-\hat{y_i})^2}{\sum(y_i-\overline{y})^2}$

where $\hat{y_i}$ is the predicted value of the i-th ground truth, and $\overline{y}$ is the average value of the gound truth.

# One Final Warning

Tonight you have seen how the input features you give to a machine learning model can be very useful in correctly predicting another output property of your data. However, **do not confuse correlation with causation.** Just because some features follow very closely to other features does not mean that one causes the other. It could be that both features are influenced by another unknown feature, or that both just happen to be similar in your data. Always be careful when making assumptions about your data and your model, and investigate further before making any definitive conclusions about the relationship in your data.

A great way to view how our model did is to use a **confusion matrix**, which lets us view how our model did in predicting the output against what it actually was.
